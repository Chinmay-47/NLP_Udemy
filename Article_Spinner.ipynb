```
from bs4 import BeautifulSoup

soup = BeautifulSoup(open('positive.review').read(),'html.parser')
positive_reviews = soup.findAll('review_text')
```

```
import nltk
nltk.download('punkt')

trigrams = {}
i = 0

for review in positive_reviews:
    s = review.text.lower()
    tokens = nltk.tokenize.word_tokenize(s)
    tokens = [token for token in tokens if len(token)>2]
    for i in range((len(tokens)-2)):
        k = (tokens[i], tokens[i+2])
        if k not in trigrams:
            trigrams[k] = []
        trigrams[k].append(tokens[i+1])
print(trigrams)
```

```
from future.utils import iteritems

trigram_probabilities = {}

for k,words in iteritems(trigrams):
    if len(set(words))>1:
        d = {}
        n=0
        for w in words:
            if w not in d:
                d[w] = 0
            d[w]+=1
            n+=1
        for w,c in iteritems(d):
            d[w] = float(c)/n
        trigram_probabilities[k] = d
```

```
trigram_probabilities
```

```
import random

def random_sample(d):
    # choose a random sample from dictionary where values are the probabilities
    r = random.random()
    cumulative = 0
    for w, p in iteritems(d):
        cumulative += p
        if r < cumulative:
            return w
```

```
def test_spinner():
    review = random.choice(positive_reviews)
    s = review.text.lower()
    print("Original:", s)
    tokens = nltk.tokenize.word_tokenize(s)
    for i in range(len(tokens) - 2):
        if random.random() < 0.2: # 20% chance of replacement
            k = (tokens[i], tokens[i+2])
            if k in trigram_probabilities:
                w = random_sample(trigram_probabilities[k])
                tokens[i+1] = w
    print("Spun:")
    print(" ".join(tokens).replace(" .", ".").replace(" '", "'").replace(" ,", ",").replace("$ ", "$").replace(" !", "!"))
```

```
if __name__ == '__main__':
    test_spinner()
```
